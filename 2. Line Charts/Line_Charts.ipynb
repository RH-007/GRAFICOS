#### **Seleccion del Dataset**

El conjunto de datos rastrea los streams diarios globales en el servicio de streaming de mñusica Spotifu. En el set de datos solo centra en 5 canciones populares del 2017 y 2018 y estas fuern:

- Shape of you, de Ed Sheeram
- Despacito, de Luis Fonsi
- Something Just Like This, de Chainsmokers y Coldplay
- HUMBLE, de Kendrick Lamar
- Unforgettable, de French Montana

```python
## Librerias
import pandas as pd
```

```python
## Importamos data

data = pd.read_csv("spotify.csv", index_col="Date", parse_dates=True)
```

```python
data.head()
```

Vemos que hay presencia de datos NaN eso quiero decir que hay datos ausentes, NaN es la abreviatura de "Not a Number". Tambien podemos echar un vistazo a las últimas cinco filas de los datos haciendo sólo un pequeño cambio, usaremos `tail()`.

```python
data.tail()
```

Afortunadamente, todo parece correcto, con millones de streams globales diarios para cada canción, y podemos preceder a trazar los datos.

## **Trazar los datos**

Ahora que el conjunto de datos está cargadp en el block de python, solo necestiamos una linea de codigo para hacer el gráfico de lineas.

```python
## Necesario cargar las librerias necesarias

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=data)
```
